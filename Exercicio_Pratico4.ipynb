```python

print("Bem vindo ao Controle de estoque Mercearia Luciano Duarte Miquelino")# Identificador pessoal
listaProdutos = []
def cadastrodeProdutos(cod):# função cadastro de produtos
    print("Cadastro de Produtos")
    print("Código do produto a ser cadastrado é: {}" .format(cod))
    nome = input("Digite o nome do produto:")
    codigo = input("Digite o código do produto:")
    fabricante = input("Digite o fabricante do produto:")
    valor = int(input("Digite o valor do produto R$:"))
    dicionarioProduto = {"cod": cod,  # dicionario
                         "nome": nome,
                         "fabricante": fabricante,}
                         
    listaProdutos.append(dicionarioProduto.copy())
def consultarProdutos():# função consulta produtos
    while True:
       try:
         print("Consulte todos os produtos")
         opConsultar = int(input("Entre com a opção desejda:\n"
                            "1 - Consultar Produto\n"
                            "2 - Consultar por Código do produto\n"
                            "3 - Consultar por Fabricante\n"
                            "4 - Retorna\n"
                            "\n>>"))
         if opConsultar == 1:
             print("Consulta todos os produtos")
             for produto in listaProdutos:
              for key, value in produto.items():
                 print("{} : {}" .format(key,value))
         elif opConsultar == 2:
             print("Consulta por código do produto")
             entrada = int(input("Digite o código do produto:"))
             for produto in listaProdutos:
              if(produto["cod"] == entrada):
                for key, value in produto.items():
                  print("{} : {}" .format(key, value))
         elif opConsultar == 3:
             print("Consulta por Fabricante")
             entrada = (input("Digite o Fabricante do produto:"))
             for produto in listaProdutos:
                 if (produto["fabricante"] == entrada):
                     for key, value in produto.items():
                         print("{} : {}" .format(key, value))
         elif opConsultar == 4:
             return
         else:
             print("Pesquisa inválida")
         continue
       except:
         break
def removerProduto():# função remover produto
    entrada = int(input("Digite o código do produto:"))
    for produto in listaProdutos:
        if produto["cod"] == entrada:
            listaProdutos.remove(produto)
            print("Produto removido com sucesso!!!")
            return
codigoProduto = 0 # lista de código para cada produto
while True:
    try:
        opcao = int(input("Digite opção desejada:\n"
                   "1-Cadastro de produtos\n"
                   "2-Consultar Produtos\n"
                   "3-Remover Produtos\n"
                   "4-Sair"
                   "\n>>"))
        if opcao == 1:
            codigoProduto = codigoProduto + 1
            cadastrodeProdutos(codigoProduto)
        elif opcao == 2:
            consultarProdutos()
        elif opcao == 3:
            removerProduto()
        elif opcao == 4:
          print("Finalizar programa")
          break
        else:
          print("Pesquisa inválida")
        continue
    except:
        break

```
